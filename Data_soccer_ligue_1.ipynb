```
import requests
from bs4 import BeautifulSoup
import pandas as pd

url='https://fbref.com/fr/comps/13/historique/Saisons-Ligue-1'
# lien de la page 

response= requests.get(url)
if response.ok:
  soup = BeautifulSoup(response.text, 'html')
  tds = soup.findAll('tr')
print(len(tds)) #verification du nombre de lignes
#(print(list(tr))for tr in tds)
```

Creation du dataframe avec les valeurs

```
trs= table.find_all('tr')
for tr in trs:
  tds= tr.find_all('td')
  row= [td.text.replace('\n','') for td in tds]
  #df= df.apply(lambda col: col.drop_duplicates().reset_index(drop=True))
  df.append(pd.Series(row,index=columns),ignore_index=True) 
print(df)

#df.to_csv('historique ligue 1.csv', index=False)
```
